# article

```python
# month
import os

path_dir = 'C:/Users/YDW/Desktop/capston/aif_Articledata/URL'
file_list = os.listdir(path_dir)
#url 파일 이름 모두 가져오기
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen  
import re

for title in file_list:
    f = open('C:/Users/YDW/Desktop/capston/aif_Articledata/URL/{}'.format(str(title)), 'r', encoding='utf-8')
    rhref = f.read().splitlines()
    f.close()
    art = []
    num = 1
    rep = []
    for url in rhref:
        print('{} / {}번째 수행중'.format(title, num))
        num=num+1
        soup = BeautifulSoup(urlopen(url).read(), 'html.parser')
        b = soup.select('article > p')
        text = ''
        for i in b:
            text = text+str(i)
            
        ntext = re.sub('[a-zA-Z0-9]|:|/|=|"|>|<|_|-|\t|\n|','',text)
        ntext = ntext.replace(u'\xa0', u' ')
        art.append(ntext)
        # article
        p = soup.find_all('div','comment_border_margin')
        reply = []
        
        for i in p:  
            np = str(i)
            np2 = np.split('\n')
            newtext = ''
            for text in np2:
                newtext = newtext+text
            reply.append(newtext)
        
        for word in reply:
            word2 = re.search(r'[>](.*)[<]', str(word))
            rep.append(word2.group(1))
        # reply
    f = open('C:/Users/YDW/Desktop/capston/aif_Articledata/Article/{}'.format(title), 'w', encoding='utf-8')
    for i in art:
        i = i+'\n'
        f.write(str(i))
    f.close()  
    
    rep = [text.strip() for text in rep]
    rep = [re.sub('[-=.#,"/)?!$};(]', '', doc) for doc in rep]
    rep = [re.sub('(br)|(<)|(>)|(/)|(-)|(_)|(!)|(\r)|(=)', '', doc) for doc in rep]
    
    f = open('C:/Users/YDW/Desktop/capston/aif_Articledata/Reply/{}'.format(str(title)), 'w', encoding='utf-8')
    for i in rep:
        i = str(i)+'\n'
        f.write(i)
    f.close()
```
